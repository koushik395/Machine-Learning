# Write a program to implement k-Nearest Neighbor algorithm to classify the data set. Print both correct and wrong predictions.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the Titanic Dataset.
## You can download the Dataset From Kaggle -> [kaggle](https://www.kaggle.com/datasets/brendan45774/test-file)

```python
titanic = pd.read_csv("tested.csv")
df = pd.DataFrame(titanic)
df.head()
```

```python
df.describe()
```

## Replacing the Gender with Binary numbers

```python
df['Sex'].replace('female',0,inplace=True)
df['Sex'].replace('male',1,inplace=True)
df
```

```python
df['Fare'].isna().sum()
```

## Replacing the missing values with mean and median.
## Here we will not replace Cabin as we dont use that feature.

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Fare'].fillna(df['Fare'].median(),inplace=True)
```

```python
df.corr()
```

```python
X = df.drop(['Survived','Name','Cabin','Embarked','Ticket'],axis=1)
y = df['Survived'].values
X
```

## Model Training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Using KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 3, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
# Predicting a new result
print(classifier.predict(sc.transform([[892,3,1,34.50000,0,0,7.8292]])))
```

## Printing Both correct and wrong predictions

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Model Evaluation

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
